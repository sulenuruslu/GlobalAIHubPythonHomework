```python
i = 0
name = str(input("Please enter your name: "))
surname = str(input("Please enter your surname: "))
print("\n")

while i < 3:
    checkname = str(input("Please enter student name: "))
    checksurname = str(input("Please enter student surname: "))
    
    if (checkname == name and checksurname == surname):
        print("Welcome", name, surname)
        print("_"*20)
        break
    else:
        i += 1
        if i == 3:
            print("Please try again later.")
            print("_"*20)
            
i=0
courselist = ["Maths", "Chemistry", "Physics", "Biology", "Litarature", "History"]
studentcourses = []
course = ""
print("Please choose your lessons from there.")
print(courselist)
print("-"*20)
  

numberofcourses = int(input("Please enter the number of courses you took: "))
if numberofcourses < 3  or numberofcourses > 5:
  print("Sorry, you failed the class.")
else:
  for i in range(numberofcourses):
    course = input("Please write your course: ")
    studentcourses.append(course.title())
    
print("You took following courses")
for i in range(numberofcourses):
  print(f"{i+1}. {studentcourses[i]}")

while True:
  course = input("Enter your course name for exam result: ")
  if course in studentcourses:
    midterm = int(input("Midterm Exam: "))
    final = int(input("Final: "))
    project = int(input("Project: "))
    print("-"*20)

    grade = midterm * 0,3 + final * 0,5 + project * 0,2
    if grade > 90:
      print("Congrats, you passed the course! Your grade is AA.")
    elif 70 < grade < 90:
      print("Congrats, you passed the course! Your grade is BB.")
    elif 50 < grade < 70:
      print("Congrats, you passed the course! Your grade is CC.")
    elif 30 < grade < 50:
      print("Congrats, you passed the course! Your grade is DD.")
    else:
      print("Sorry, you failed the exam. Your grade is FF")
  else:
    print("This course is not named in your course list.")

#Hocam merhaba, ben pythonda çok yeniyim hatta sizin sayenizde tanıştım diyebilirim. 
#Teşekkür ederim her şey için fazlasıyla ilerleme kaydettim sayenizde. 
#Düzeltemediğim bir hata alıyorum bu şekilde attığım için özür dilerim.
```
